Higher chest pain type (cp) and max heart rate (thalach) â†’ Higher heart disease risk

## Feature Engineering - Normalize Numerical Columns

```python
scaler = StandardScaler()
x=df.drop(columns=['target'])
y=df['target']
x_scaled=scaler.fit_transform(x)
```

## Split Data

```python
x_train, x_test, y_train, y_test = train_test_split(x_scaled, y, test_size=0.2, random_state=42)
```

## Train Model

```python
model= RandomForestClassifier(n_estimators=100, random_state=42)
model.fit(x_train, y_train)
```

### Prediction

```python
y_pred = model.predict(x_test)
```

### # Model Evaluation

```python
print("Classification Report:")
print(classification_report(y_test, y_pred))
print("Accuracy Score:" , accuracy_score(y_test, y_pred))
```

```python
# Explainable AI using SHAP
explainer = shap.Explainer(model, x_train)
shap_values = explainer(x_test, check_additivity=False)
shap.summary_plot(shap_values, x_test, feature_names=x.columns)
```

```python
# Automated Report
summary = f"""
Heart Disease Prediction Report:

- Model: RandomForestClassifier
- Accuracy: {accuracy_score(y_test, y_pred):.2f}
- Key Features Influencing Predictions: SHAP Summary Plot
"""
print(summary)
```
